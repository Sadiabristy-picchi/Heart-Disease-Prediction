**Likelihood :**

P(X= x0,x1,x2 | Y=y)

= P(X=x0|Y=y) *P(X=x1|Y=y) * P(X=x2|Y=y)

**Likelihood** refers to the probability of observing a set of data, given a particular parameter or hypothesis

**The prior** probability distribution can be represented by a probability density function, which assigns probabilities to all possible values of the parameter or hypothesis.

**The probability density function (PDF) of the normal distribution formula:**

f(x | μ, σ^2) = (1 / (σ * √(2π))) * exp(-(x - μ)^2 / (2 * σ^2))

```python
from sklearn.model_selection import train_test_split
newDF = df.copy()
x = newDF[['age', 'sex', 'cp', 'chol', 'fbs','thalach', 'target']]
y = newDF[['target']]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.15, random_state = 0)
print(x_train)
print(y_train)
print(x_test)
print(y_test)
```

```python
def naive_bayeisian(x_train,age,sex,cp,chol,fbs,thalach,target):
 #For Disease : 1
 #prior probability
 prior =x_train[x_train['target']==1].count()[0] / len(x_train)

 #discreet feature
 #likelihood

 sex1 = x_train[(x_train['sex']==sex) & (x_train['target']==1)].count()[0] / x_train[x_train['target']==1].count()[0]
 cp1= x_train[(x_train['cp']==cp) & (x_train['target']==1)].count()[0] / x_train[x_train['target']==1].count()[0]
 fbs1= x_train[(x_train['fbs']==fbs) & (x_train['target']==1)].count()[0] / x_train[x_train['target']==1].count()[0]

#continuous feature
#likelihood

 age_mean1 = x_train[x_train['target'] == 1]['age'].mean()
 age_std1 = x_train[x_train['target'] == 1]['age'].std()
 age1 = (1 / (np.sqrt(2 * np.pi) * age_std1)) * np.exp(-((age - age_mean1) ** 2 / (2 * age_std1 ** 2)))
 
 chol_mean1 = x_train[x_train['target'] == 1]['chol'].mean()
 chol_std1 = x_train[x_train['target'] == 1]['chol'].std()
 chol1 = (1 / (np.sqrt(2 * np.pi) * chol_std1)) * np.exp(-((chol - chol_mean1) ** 2 / (2 * chol_std1 ** 2)))

 thalach_mean1 = x_train[x_train['target'] == 1]['thalach'].mean()
 thalach_std1 = x_train[x_train['target'] == 1]['thalach'].std()
 thalach1 = (1 / (np.sqrt(2 * np.pi) * thalach_std1)) * np.exp(-(( thalach - thalach_mean1) ** 2 / (2 * thalach_std1 ** 2)))

 #Posterior probability for 1
 target_1 = sex1 * cp1 *fbs1 * age1 *  chol1 * thalach1 * prior

#For Disease: 0
#calculate prior probability
 prior2 = x_train[x_train['target']==0].count()[0] / len(x_train)

 sex = x_train[(x_train['sex']==sex) & (x_train['target']==0)].count()[0] / x_train[x_train['target']==0].count()[0]
 cp= x_train[(x_train['cp']==cp) & (x_train['target']==0)].count()[0] / x_train[x_train['target']==0].count()[0]
 fbs= x_train[(x_train['fbs']==fbs) & (x_train['target']==0)].count()[0] /x_train[x_train['target']==0].count()[0]

#continuous feature
# likelihood 

 age_mean = x_train[x_train['target'] == 0]['age'].mean()
 age_std = x_train[x_train['target'] == 0]['age'].std()
 age = (1 / (np.sqrt(2 * np.pi) * age_std)) * np.exp(-((age - age_mean) ** 2 / (2 * age_std** 2)))
 
 chol_mean = x_train[x_train['target'] == 0]['chol'].mean()
 chol_std = x_train[x_train['target'] == 0]['chol'].std()
 chol = (1 / (np.sqrt(2 * np.pi) * chol_std)) * np.exp(-((chol - chol_mean) ** 2 / (2 * chol_std ** 2)))

 thalach_mean = x_train[x_train['target'] == 0]['thalach'].mean()
 thalach_std = x_train[x_train['target'] == 0]['thalach'].std()
 thalach = (1 / (np.sqrt(2 * np.pi) * thalach_std)) * np.exp(-(( thalach -  thalach_mean) ** 2 / (2 *  thalach_std ** 2)))

#Posterior probability for 0
 target_0 = sex * cp * fbs * age *  chol * thalach * prior2

 if target_1 < target_0:
    return 0
 else:
    return 1
```

```python
target_pred = []

for i, data in x_test.iterrows():
  age = data['age']
  sex = data['sex']
  cp = data['cp']
  chol = data['chol']
  fbs = data['fbs']
  thalach = data['thalach']
  target = data['target']

  pred = naive_bayeisian(x_train,age,sex,cp,chol,fbs,thalach,target)
  target_pred.append(pred)
```

```python
from sklearn.metrics import confusion_matrix
y_test_arr = y_test['target'].values
y_pred = np.array(target_pred)
con_mat = confusion_matrix(y_test_arr, y_pred)
print(con_mat)
```

# **Single Layer Perceptron**

```python
#equation
def f(X, W):
  return W[0] + np.dot(X, W[1:])

#activation function
def A(x):
  if x>0:
    return 1
  else:
    return 0

#classifier
def Y(X, W):
  return A(f(X,W))
```

```python
def train(x_train, y_train):
  W = np.zeros(x_train.shape[1] + 1) #initializing weights
  eta = 0.05 #learning rate
  epoch = 0
  prev_mse = np.inf
  best_weights = None

  while True:
    target_pred_p = []
    for i in range(len(x_train)):
      X = x_train.iloc[i].values
      Ti = y_train[i][0]
      Pi = Y(X, W)
      target_pred_p.append(Pi)
      delw = eta * (Ti - Pi)
      W[0] += delw * 1 #Update bias weight
      W[1:] += delw * X #Update feature weight
    
    #Mean squared error
    
    MSE = np.mean((np.array(target_pred_p) - y_train) ** 2)

    epoch += 1

    print(f"Epoch {epoch}: MSE = {MSE}")
    if epoch > 1 and MSE >= prev_mse:
      break
    prev_mse = MSE

    if best_weights is None or MSE<best_mse:
      best_weights = np.copy(W)
      best_mse = MSE

  return W, MSE
```

```python
def test_accuracy(x_test, y_test, W):
  accurate_preds = 0
  for i in range(len(x_test)):
    X = x_test.iloc[i].values
    Ti = y_test[i][0]
    Pi = Y(X, W)
    if Pi == 1 and Ti == 1:
      accurate_preds += 1
    elif Pi == 0 and Ti == 0:
      accurate_preds += 1
    
  accuracy = accurate_preds/len(x_test)
  return accuracy
```

```python
x_train = pd.DataFrame(x_train) 
x_test = pd.DataFrame(x_test)  
y_train = np.array(y_train).reshape(-1, 1) 
y_test = np.array(y_test).reshape(-1, 1) 
```

```python
weight, train_mse = train(x_train, y_train)
accuracy_p = test_accuracy(x_test, y_test, weight)
acc_pre = accuracy_p * 100
print(f"Train MSE: {train_mse}")
print(f"Test Accuracy: {acc_pre}%")
print(f"Learned Weights: {weight}")
```
